```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape a recipe from the provided URL
def scrape_recipe(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        recipe_name = soup.find("h1",class_="recipe-title").text
        ingredients_list =  [item.text.strip() for item in soup.find_all("li", itemprop="ingredients")]
        method = [item.text.strip() for item in soup.find_all("li", itemprop="recipeInstructions")]
        serves = soup.find("span",itemprop="recipeYield").text
        time = soup.find("span",itemprop="totalTime").text

        recipe_data = {
            "Recipe Name": recipe_name,
            "Total Time": time,
            "Ingredients": ", ".join(ingredients_list),
            "Method": method,
            "Serves": serves
        }

        return recipe_data
    else:
        print(f"Failed to retrieve data from the website. Status code: {response.status_code}")
        return None

recipe_urls = [
    "https://www.archanaskitchen.com/pudina-tambuli-recipe",
    "https://www.archanaskitchen.com/jowar-atta-roti-recipe-sorghum-wheat-phulka-recipe",
    "https://www.archanaskitchen.com/mangalorean-bhende-puli-kodel-recipe-okra-in-tamarind-gravy",
    "https://www.archanaskitchen.com/stuffed-gilka-recipe-maharashtrian-style-stuffed-silk-squash",
    "https://www.archanaskitchen.com/thengai-saadam-spiced-coconut-rice-recipe",
    "https://www.archanaskitchen.com/peanut-curry-recipe",
    "https://www.archanaskitchen.com/menthe-fenugreek-seeds-tambli-recipe-karnataka-recipes",
    "https://www.archanaskitchen.com/vendakka-stew-recipe-kerala-style-okra-stew",
    "https://www.archanaskitchen.com/malvani-kurkuri-bhindi-recipe",
    "https://www.archanaskitchen.com/coorg-style-kumbala-curry-recipe-pumpkin-sabzi",
    "https://www.archanaskitchen.com/beetroot-sesame-thepla-recipe",
    "https://www.archanaskitchen.com/goan-pumpkin-sabzi-recipe",
    "https://www.archanaskitchen.com/carrot-sweet-potato-mushrooms-fritters-recipe",
    "https://www.archanaskitchen.com/classic-andhra-style-chicken-curry-recipe",
    "https://www.archanaskitchen.com/chettinadu-style-potato-kaarakozhambu-recipe",
    "https://www.archanaskitchen.com/bulgur-wheat-stuffed-mini-peppers-oven-roasted-mini-peppers-stuffed-with-quick-cooking-form-of-wheat-with-roasted-spices",
    "https://www.archanaskitchen.com/andhra-style-vankaya-kothimeera-karam-koora-recipe",
    "https://www.archanaskitchen.com/cherupayar-thoran-recipe-green-gram-dry-sabzi-kerala-style",
    "https://www.archanaskitchen.com/gutti-dondakaya-vepudu-recipe-andhra-style-stuffed-ivy-gourds",
    "https://www.archanaskitchen.com/goan-style-khatkhatem-recipe-mixed-vegetable-gravy",
    "https://www.archanaskitchen.com/tamil-nadu-style-broccoli-poriyal-recipe-broccoli-stir-fry-recipe",
    "https://www.archanaskitchen.com/kuvale-sasam-recipe-konkani-ash-gourd-curry-no-onion-no-garlic",
    "https://www.archanaskitchen.com/besan-aur-mirch-ki-sabzi-recipe",
    "https://www.archanaskitchen.com/kathirikai-urulaikizhanghu-pachidi-recipe-brinjals-and-potatoes-with-moong-dal",
    "https://www.archanaskitchen.com/jolada-roti-jowar-bhakri-recipe",
    "https://www.archanaskitchen.com/creamy-veg-spring-roll-recipe-kids-recipes-made-with-del-monte",
    "https://www.archanaskitchen.com/karnataka-style-sandige-huli-recipe-toor-dal-dumplings-in-gravy-recipe",
    "https://www.archanaskitchen.com/andhra-style-palak-kura-pappu-recipe-spinach-dal-recipe",
    "https://www.archanaskitchen.com/chettinad-style-pepper-mushroom-recipe",
    "https://www.archanaskitchen.com/lilve-ki-sabzi-recipe-green-chickpea-curry",
    "https://www.archanaskitchen.com/gutti-vankaya-recipe-andhra-style-stuffed-brinjal",
    "https://www.archanaskitchen.com/mangalorean-style-bendekayi-huli-koddel-recipe-ladies-finger-dal-without-coconut-recipe",
    "https://www.archanaskitchen.com/allahabadi-tehri-recipe-up-style-vegetable-rice",
    "https://www.archanaskitchen.com/hyderabad-style-kuska-recipe-biryani-rice-recipe",
    "https://www.archanaskitchen.com/mixed-sprouts-in-chettinad-masala-recipe",
    "https://www.archanaskitchen.com/himachal-style-pahari-sidu-recipe-stuffed-bread-recipe",
    "https://www.archanaskitchen.com/bengaluru-style-brinjal-gravy-recipe-for-biryani",
    "https://www.archanaskitchen.com/narthangai-curry-sweet-and-sour-citron-fruit-curry",
    "https://www.archanaskitchen.com/chole-aloo-tikki-chaat-an-indian-street-food-favorite",
    "https://www.archanaskitchen.com/bajra-aloo-paratha-recipe",
    "https://www.archanaskitchen.com/video-recipe-vada-sambar-non-fried-lentil-fritters-with-lentil-curry",
    "https://www.archanaskitchen.com/chettinad-keerai-mandi-recipe-with-amaranth-greens",
    "https://www.archanaskitchen.com/goan-style-empada-de-camarao-recipe-prawn-pie",
    "https://www.archanaskitchen.com/chettinad-special-steamed-lentil-dumplings-gravy-kola-urundai-kuzhambu",
    "https://www.archanaskitchen.com/mangalorean-style-vaingana-mashinga-sange-sukhe-sukka-recipe-dry-brinjal-and-cashewnut-stir-fry-recipe",
    "https://www.archanaskitchen.com/karnataka-style-orange-peels-curry-recipe",
    "https://www.archanaskitchen.com/kodava-kummu-curry-recipe-coorg-style-mushroom-curry",
    "https://www.archanaskitchen.com/sweet-potato-broccoli-and-tofu-curry-recipe",
    "https://www.archanaskitchen.com/bharli-vangi-recipe-karwar-nkgsb-style-brinjal-curry",
    "https://www.archanaskitchen.com/mughalai-shahi-kaju-aloo-ki-sabzi-recipe",
    "https://www.archanaskitchen.com/capsicum-in-coconut-and-peanut-gravy-recipe",
    "https://www.archanaskitchen.com/baked-matar-kachori-recipe-delicious-evening-snack",
    "https://www.archanaskitchen.com/chettinad-kathirikai-chops-recipe-brinjal-curry-for-biryanis",
    "https://www.archanaskitchen.com/andhra-style-sorakaya-kura-recipe-bottle-gourd-and-toor-dal-recipe",
    "https://www.archanaskitchen.com/vendakkai-poriyal-recipe-south-indian-style-okra-stir-fry-spiced-with-sambar-powder",
    "https://www.archanaskitchen.com/hyderabadi-shahi-mixed-vegetable-curry-recipe",
    "https://www.archanaskitchen.com/hyderabadi-khadi-masoor-ki-dal-hyderabadi-style-lentil-curry",
    "https://www.archanaskitchen.com/karnataka-style-mixed-vegetable-saagu-recipe-spicy-vegetable-coconut-gravy-recipe",
    "https://www.archanaskitchen.com/best-punjabi-kadhi-pakora-spicy-yogurt-curry-with-spinach-dumplings",
    "https://www.archanaskitchen.com/goan-style-mushroom-vindaloo-recipe",
    "https://www.archanaskitchen.com/kashmiri-munj-haakh-recipe-knol-khol-flavored-with-asafoetida",
    "https://www.archanaskitchen.com/hyderabadi-bagare-tamatar-spiced-tossed-tomato-curry",
    "https://www.archanaskitchen.com/milagai-bajji-stuffed-with-masala-chicken-recipe",
    "https://www.archanaskitchen.com/mutton-chaap-recipe",
    "https://www.archanaskitchen.com/masoor-dal-ghassi-recipe-lentils-in-tangy-coconut-curry",
    "https://www.archanaskitchen.com/rajasthani-laal-maas-recipe-mutton-in-red-spicy-gravy",
    "https://www.archanaskitchen.com/goan-style-black-chickpeas-and-cucumber-masala-taushem-tondak-recipe",
    "https://www.archanaskitchen.com/oats-peas-tofu-stuffed-paratha-recipe",
    "https://www.archanaskitchen.com/bhaat-na-rasawala-muthia-dhokla-recipe-gujarati-rasiya-dhokla",
    "https://www.archanaskitchen.com/odisha-style-dal-tadka-recipe",
    "https://www.archanaskitchen.com/pavakka-pachadi-recipe-bitter-gourd-pachadi",
    "https://www.archanaskitchen.com/goan-vegetable-curry-recipe",
    "https://www.archanaskitchen.com/sindhi-koki-sindhi-style-flaky-flatbread-with-onions",
    "https://www.archanaskitchen.com/aloo-konir-dom-recipe-assamese-egg-potato-curry",
    "https://www.archanaskitchen.com/thekera-tenga-recipe",
    "https://www.archanaskitchen.com/bibya-usal-recipe-karwar-spiced-cashew-nut-sabzi",
    "https://www.archanaskitchen.com/urulai-kizhangu-masiyal-potatoes-in-a-tangy-lemon-and-ginger-gravy",
    "https://www.archanaskitchen.com/stuffed-bhindi-with-paneer",
    "https://www.archanaskitchen.com/red-lentil-curry-without-coconut-recipe",
    "https://www.archanaskitchen.com/methi-mooli-makkai-paratha-recipe",
    "https://www.archanaskitchen.com/padipe-saasmi-recipe-udupi-style-amaranth-leaves-curry-with-coconut-tamarind",
    "https://www.archanaskitchen.com/jain-style-dal-makhani-recipe",
    "https://www.archanaskitchen.com/paya-curry-recipe-lamb-trotters",
    "https://www.archanaskitchen.com/restaurant-style-phool-gobhi-masala-recipe",
    "https://www.archanaskitchen.com/andhra-style-davva-aava-koora-recipe-sweet-and-spicy-banana-stem-peanut-curry-recipe",
    "https://www.archanaskitchen.com/goan-style-recheado-bangdo-recipe-stuffed-mackerel",
    "https://www.archanaskitchen.com/methi-kadhi-fenugreek-leaf-yogurt-curry",
    "https://www.archanaskitchen.com/masala-vada-curry-recipe-chana-dal-vada-curry-recipe",
    "https://www.archanaskitchen.com/bihari-shakarkand-bhujiya-recipe",
    "https://www.archanaskitchen.com/vali-bhajji-ambat-recipe-konkan-style-spiced-malabar-spinach-dal-recipe",
    "https://www.archanaskitchen.com/rava-kakdi-thalipeeth-recipe-maharashtrian-style-semolina-cucumber-flatbread",
    "https://www.archanaskitchen.com/konkani-style-alsande-randayi-recipe-black-eyed-beans-and-yam-gravy",
    "https://www.archanaskitchen.com/kaju-karela-recipe-bitter-gourd-and-cashew-nuts-stir-fry",
    "https://www.archanaskitchen.com/foxtail-millet-lemon-rice-recipe-thenai-sadam",
    "https://www.archanaskitchen.com/mangalorean-style-kuvalyacho-pollav-recipe-yellow-pumpkin-curry-recipe",
    "https://www.archanaskitchen.com/ven-pongal-south-indian-rice-and-lentil-pudding",
    "https://www.archanaskitchen.com/achaari-aloo-gobi-recipe-cauliflower-in-pickling-spices-indian-vegetarian",
    "https://www.archanaskitchen.com/andhra-style-tomato-cauliflower-koora",
    "https://www.archanaskitchen.com/rawal-pindi-style-chole-recipe",
    "https://www.archanaskitchen.com/andhra-style-vendakaya-charu-recipe-ladies-finger-rasam-recipe",
    "https://www.archanaskitchen.com/andhra-style-pindi-miriyam-recipe",
    "https://www.archanaskitchen.com/karuvepilai-chicken-fry-recipe",
    "https://www.archanaskitchen.com/paneer-gassi-recipe",
    "https://www.archanaskitchen.com/aloo-bhindi-ki-sabzi-spicy-potatoes-and-lady-s-finger",
    "https://www.archanaskitchen.com/whole-wheat-spinach-paratha-recipe",
    "https://www.archanaskitchen.com/kashmiri-rogan-josh-recipe",
    "https://www.archanaskitchen.com/mangalorean-style-lobia-gazzi-recipe-black-eyed-peas-in-coconut-curry-recipe",
    "https://www.archanaskitchen.com/padpe-uppukari-harive-soppu-palya-recipe",
    "https://www.archanaskitchen.com/grilled-paneer-methi-tikki-recipe",
    "https://www.archanaskitchen.com/bhugi-dal-paratha-recipe-dal-jo-phulko",
    "https://www.archanaskitchen.com/udid-methi-recipe-karwar-style-green-mango-curry-with-methi",
    "https://www.archanaskitchen.com/brinjal-capsicum-sabji-recipe",
    "https://www.archanaskitchen.com/mushroom-pulav",
    "https://www.archanaskitchen.com/karate-batate-puddi-sagale-recipe-konkani-style-bitter-gourd-and-potato-curry",
    "https://www.archanaskitchen.com/malvani-kala-chana-masala-recipe-spicy-kala-chana-curry",
    "https://www.archanaskitchen.com/mangalorean-manoli-curry-recipe-indian-gherkins-curry",
    "https://www.archanaskitchen.com/raw-tomato-curry-recipe-hare-tamatar-ki-sabzi",
    "https://www.archanaskitchen.com/thengai-poornam-kozhukattai-ukadiche-modak-recipe-with-rice-flour",
    "https://www.archanaskitchen.com/cauliflower-curry-with-bell-peppers-and-carrots-in-spiced-milk-gravy",
    "https://www.archanaskitchen.com/kacha-kela-bhindi-recipe",
    "https://www.archanaskitchen.com/vankaya-kothimeera-karam-recipe-andhra-style-brinjals-cooked-with-spicy-coriander-mix",
    "https://www.archanaskitchen.com/kottambari-tambuli-recipe-udupi-style-raw-coriander-kadhi",
    "https://www.archanaskitchen.com/sambar-powder-podi",
    "https://www.archanaskitchen.com/stuffed-bhindi-recipe-with-peanuts",
    "https://www.archanaskitchen.com/konkani-style-southe-koddel-recipe-mangalore-cucumber-in-coconut-curry-recipe",
    "https://www.archanaskitchen.com/prawns-biryani-recipe",
    "https://www.archanaskitchen.com/kadai-chole-recipe-dry-chickpeas-with-capsicum",
    "https://www.archanaskitchen.com/poosanikai-rasavangi-recipe-ash-gourd-curry-with-dal",
    "https://www.archanaskitchen.com/kerala-style-puliserry-mor-kozhambu",
    "https://www.archanaskitchen.com/south-indian-style-sorakkai-kootu-recipe-bottle-gourd-cooked-with-lentils-recipe",
    "https://www.archanaskitchen.com/bendekayi-koddel-bolu-huli-recipe-mangalorean-style-bhindi-sambar-recipe",
    "https://www.archanaskitchen.com/potato-and-peas-in-coconut-milk-curry-recipe",
    "https://www.archanaskitchen.com/shahi-egg-curry-recipe-in-tomato-cashew-gravy",
    "https://www.archanaskitchen.com/chettinad-prawn-curry-recipe",
    "https://www.archanaskitchen.com/coconut-burfi-kopra-pak-fresh-coconut-fudge",
    "https://www.archanaskitchen.com/maharashtrian-bakar-bhaji-recipe",
    "https://www.archanaskitchen.com/goan-poee-recipe-delicious-whole-wheat-buns-recipe",
    "https://www.archanaskitchen.com/soya-methi-palak-ki-sabzi-recipe",
    "https://www.archanaskitchen.com/goan-style-mushroom-shagoti-recipe",
    "https://www.archanaskitchen.com/assamese-machor-tenga-anja-recipe-fish-curry-with-kokum",
    "https://www.archanaskitchen.com/quick-honeyed-paneer",
    "https://www.archanaskitchen.com/mutton-chaap-recipe",
    "https://www.archanaskitchen.com/masoor-dal-ghassi-recipe-lentils-in-tangy-coconut-curry",
    "https://www.archanaskitchen.com/rajasthani-laal-maas-recipe-mutton-in-red-spicy-gravy",
    "https://www.archanaskitchen.com/mangalorean-chana-gassi-recipe",
    "https://www.archanaskitchen.com/arbi-ki-khatta-subzi-recipe-colocasia-in-yoghurt-curry",
    "https://www.archanaskitchen.com/gobi-musallam-recipe-spicy-cauliflower-creamy-gravy",
    "https://www.archanaskitchen.com/pineapple-rasam-recipe",
    "https://www.archanaskitchen.com/pita-saga-bati-basa-recipe"


]

recipes_data = []

for url in recipe_urls:
    recipe_data = scrape_recipe(url)
    if recipe_data:
        recipes_data.append(recipe_data)

df = pd.DataFrame(recipes_data)

df.to_csv('/content/drive/MyDrive/Pubmed/all_recipes.csv', index=False)

print("Data for all recipes has been scraped and saved to 'all_recipes.csv'.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Define the dish you want to find similar dishes for
target_dish = "Masoor Dal Gassi Recipe - Lentils in Tangy Coconut Curry"  # Replace with the dish name you want


# Create a CountVectorizer to convert ingredients into a bag of words
count_vectorizer = CountVectorizer()

# Convert lists in 'Method' column to strings
df['Method'] = df['Method'].apply(lambda x: ' '.join(x))

# Fit the CountVectorizer and create the ingredient matrix
ingredient_matrix = count_vectorizer.fit_transform(df['Method'])

# Calculate cosine similarity
cosine_sim = cosine_similarity(ingredient_matrix)


# Get the index of the target dish
target_index = df[df['Recipe Name'] == target_dish].index[0]

# Calculate similarity scores for all dishes
similarity_scores = list(enumerate(cosine_sim[target_index]))

# Sort the dishes based on similarity
sorted_similar_dishes = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

# Get the top 2 similar dishes
top_similar_dishes = sorted_similar_dishes[1:4]  # Exclude the target dish itself

# Extract and print the names of the top similar dishes
print("Similarity based on the ingredients used (top 3)")
print("")
for dish_index, similarity_score in top_similar_dishes:
    if similarity_score > 0.9:
      print(f"Dish: {df['Recipe Name'][dish_index]}, Probably the same dish")
    elif similarity_score > 0.6:
      print(f"Dish: {df['Recipe Name'][dish_index]}, Similarity: {similarity_score:.2f}")
```

```python
from transformers import pipeline, DistilBertTokenizer, DistilBertForQuestionAnswering
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Pubmed/all_recipes.csv')
# Define the question-answering pipeline using DistilBERT
qa_pipeline = pipeline(
    "question-answering",
    model=DistilBertForQuestionAnswering.from_pretrained("distilbert-base-cased-distilled-squad"),
    tokenizer=DistilBertTokenizer.from_pretrained("distilbert-base-cased-distilled-squad"),
)

# Define a function to get an answer based on a question
def get_answer(question, context):
    result = qa_pipeline(question=question, context=context)
    return result["answer"]

# Example usage
target_dish_index = 0  # Replace with the index of the dish you want to use as context
target_dish_name = df["Recipe Name"][target_dish_index]
context = df["Method"][target_dish_index]

# Sample questions
questions = ["What is the main ingredient?", "How to cook?", "Any tips for making it better?"]

# Print answers for each question
for question in questions:
    answer = get_answer(question, context)
    print(f"Question: {question}")
    print(f"Answer: {answer}")
    print("-" * 50)
```

```python
import spacy
import os
from torchvision import transforms
from torchvision.datasets import FashionMNIST
# Load a pre-trained spaCy model (e.g., en_core_web_sm)
nlp = spacy.load("en_core_web_sm")

# Example function for entity relation extraction
def extract_entities_relations(text):
    doc = nlp(text)

    # Extract entities and their relationships based on your use case
    cuisines = [ent.text for ent in doc.ents if ent.label_ == "CUISINE"]
    ingredients = [ent.text for ent in doc.ents if ent.label_ == "INGREDIENT"]

    # Assuming ROOT is the recipe name
    recipe_name = [token.text for token in doc if token.dep_ == "ROOT"]

    return recipe_name, cuisines, ingredients
```

```python
# Example function for question-answering using spaCy entities
def get_answer(question, context):
    # Use the entity relation extraction model to preprocess the text and extract entities
    recipe_name, cuisines, ingredients = extract_entities_relations(context)

    # Incorporate entities into the context
    enriched_context = f"{context} {' '.join(recipe_name)} {' '.join(cuisines)} {' '.join(ingredients)}"

    # Use a question-answering model for generating answers
    qa_pipeline = pipeline(
        "question-answering",
        model=DistilBertForQuestionAnswering.from_pretrained("distilbert-base-cased-distilled-squad"),
        tokenizer=DistilBertTokenizer.from_pretrained("distilbert-base-cased-distilled-squad"),
    )

    # Get the answer using the question-answering model
    result = qa_pipeline(question=question, context=enriched_context)
    return result["answer"]
```

```python
# Example usage
target_dish_index = 1  # Replace with the index of the dish you want to use as context
target_dish_name = df["Recipe Name"][target_dish_index]
context = df["Method"][target_dish_index]

# Sample questions
questions = ["What is the main ingredient?", "How to cook?", "Any tips for making it better?"]

# Print answers for each question
for question in questions:
    answer = get_answer(question, context)
    print(f"Question: {question}")
    print(f"Answer: {answer}")
    print("-" * 50)
```

```python
import csv
from collections import defaultdict

# Define the file path of the CSV file
file_path = "main_ingredient.csv"

# Define a defaultdict to store the frequency of each element
freq_dict = defaultdict(list)

# Read the CSV file and populate the defaultdict
with open(file_path, "r") as csv_file:
    csv_reader = csv.reader(csv_file)

    # Assume the header is present
    header = next(csv_reader)

    # Populate the defaultdict with frequency counts
    for row in csv_reader:
        freq_dict[row[header.index("Main Ingridient")]].append(row)

# Identify the elements with duplicate frequency and print the corresponding rows
for element, freq in freq_dict.items():
    if len(freq) > 1:
        print(f"Ingredient '{element}' occurs {len(freq)} times. The corresponding rows are:")
        for row in freq:
            print(row)
        print("-" * 80)
```
