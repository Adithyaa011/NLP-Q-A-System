```python
import requests
import os
from bs4 import BeautifulSoup

def scrape_page(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract abstract content
        try:
            abstract = soup.find('div', {'class': 'abstract', 'id': 'abstract'}).text.strip()
        except AttributeError:
            abstract = " "

        # Extract references content
        try:
            references_list = []
            references_container = soup.find('ol', {'id': 'top-references-list-1', 'class': 'references-list'})
            for li in references_container.find_all('li'):
                references_and_notes_list = li.find('ol', {'class': 'references-and-notes-list'})
                if references_and_notes_list:
                    references_text = ''.join([content.text.strip() for content in references_and_notes_list.contents if content.name != 'a'])
                    references_list.append(references_text)
            if not references_list:
                references_list = [" "]
        except AttributeError:
            references_list = [" "]

        return abstract, references_list

    else:
        # Print an error message if the request was not successful
        print(f"Error: Unable to fetch the page (Status code: {response.status_code})")
        return None, None

# Example usage
input_string = """
37968517
37937014
37886146
37886142
37870025
37860117
37787275
37773328
37770236
37765396
37763282
37753020
37751403
37737272
37694023
37670194
37655435
37630520
37602607
37599978
37599967
37587287
37567911
37507040
37469923
37417133
37390717
37383807
37304809
37292324
37253857
37244779
37203030
37196663
37167346
37152300
37144025
37103704
37102259
37026225
37019298
36983367
36892735
36888981
36814168
36797613
36778101
36770859
36732777
36729674
36727968
36727364
36727325
36707285
36696062
36677046
36670131
36643148
36606068
36588211
36515030
36510955
36432888
36412415
36412405
36370662
36285040
36254950
36149164
36110452
36012832
35993293
35918957
35867658
35853424
35844309
35836568
35813800
35813228
35791141
35791139
35773779
35677004
35655905
35654349
35647959
35645522
35627178
35578854
35504935
35502083
35370372
35353251
35339696
35326058
35326031
35325996
35324258
35315292
35291347
"""
pmid_url = []
for i in input_string:
  pmid_url.append('https://pubmed.ncbi.nlm.nih.gov/' + str(i) + '/')

count = 1
output_folder = '/content/drive/MyDrive/NLP Pubmed/Abstract'  # Set the output folder path
os.makedirs(output_folder, exist_ok=True)

for url in pmid_url:
  print("tulasi - {}".format(count))
  abstract, _ = scrape_page(url)

  if abstract is not None:
    new_abs = abstract.replace("  ", "").replace("\n\n", "\n")
    print("Abstract:", new_abs)
  else:
    print("Error: Unable to fetch the page")


  # Adding in a text file
  abstract_filename = os.path.join(output_folder, f'tulasi_{str(count).zfill(3)}A.txt')

  with open(abstract_filename, 'w') as abstract_file:
      abstract_file.write(new_abs)

  count += 1
```

```python
import os
import fitz
import pandas as pd
import re
import warnings

def read_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return ""

def sanitize_text(text):
    # Define a regular expression to remove illegal characters
    illegal_char_pattern = re.compile(r'[\x00-\x1F\x7F-\x9F]')

    # Remove or replace illegal characters
    sanitized_text = illegal_char_pattern.sub('', text)

    return sanitized_text

def create_excel_sheet(root_folder):
    # Ignore warnings during Excel sheet creation
    warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")

    # Define folder names
    folders = ['Abstract', 'References', 'Full_Text']

    # Initialize DataFrame
    data = {'Serial Number': [], 'Paper Title': [], 'Abstract': [], 'Full Text': [], 'References': []}
    df = pd.DataFrame(data)

    # Get the list of PDF files in the 'Research_Papers' folder
    pdf_files = [file for file in os.listdir(os.path.join(root_folder, 'Research_Papers')) if file.endswith('.pdf')]

    # Iterate through PDF files and extract information
    for count, pdf_file in enumerate(sorted(pdf_files), start=1):
        # Extract information from text files
        abstract_path = os.path.join(root_folder, 'Abstract', f'pepper_{str(count).zfill(3)}A.txt')
        references_path = os.path.join(root_folder, 'References', f'pepper_{str(count).zfill(3)}R.txt')
        full_text_path = os.path.join(root_folder, 'Full_Text', f'pepper_{str(count).zfill(3)}F.txt')

        abstract = sanitize_text(read_text_file(abstract_path))
        references = sanitize_text(read_text_file(references_path))
        full_text = sanitize_text(read_text_file(full_text_path))

        # Get the paper title from the PDF metadata
        pdf_document = fitz.open(os.path.join(root_folder, 'Research_Papers', pdf_file))

        pdf_document.close()

        # Append data to the DataFrame
        df = df.append({
            'Serial Number': count,
            'Paper Title': "Pepper",
            'Abstract': abstract,
            'Full Text': full_text,
            'References': references
        }, ignore_index=True)

    # Save DataFrame to Excel file
    excel_file_path = os.path.join(root_folder, 'NLP_Pubmed_Data.xlsx')
    df.to_excel(excel_file_path, index=False)

    print(f'Excel sheet created and saved to: {excel_file_path}')

if __name__ == "__main__":
    # Define the root folder path
    root_folder = '/content/drive/MyDrive/NLP Pubmed'

    # Create Excel sheet
    create_excel_sheet(root_folder)
```

```python
import os

import nltk
nltk.download('stopwords')
nltk.download('punkt')

import fitz
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def read_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return ""

def create_entity_file(root_folder, output_file_path, metadata_file_path):
    stop_words = set(stopwords.words('english'))

    with open(output_file_path, 'w', encoding='utf-8') as output_file, \
         open(metadata_file_path, 'w', encoding='utf-8') as metadata_file:

        for count in range(1, 101):
            # Extract information from text files
            full_text_path = os.path.join(root_folder, 'Full_Text', f'olive_oil_{str(count).zfill(3)}F.txt')
            abstract_path = os.path.join(root_folder, 'Abstract', f'olive_oil_{str(count).zfill(3)}A.txt')

            full_text = read_text_file(full_text_path)
            abstract = read_text_file(abstract_path)

            # Extract entities from the full text
            words = word_tokenize(full_text)
            words = [word.lower() for word in words if word.isalnum() and word.lower() not in stop_words]

            # Write entities to the output file
            output_file.write(f'Paper {count} Entities:\n')
            output_file.write(', '.join(words) + '\n\n')

            # Write metadata to the metadata file
            metadata_file.write(f'Paper {count} Metadata:\n')
            metadata_file.write(f'Title: {extract_title_from_pdf(root_folder, count)}\n')
            metadata_file.write(f'Abstract: {abstract}\n\n')

def extract_title_from_pdf(root_folder, count):
    pdf_file_path = os.path.join(root_folder, 'Papers', f'olive_oil_{str(count).zfill(3)}P.pdf')
    pdf_document = fitz.open(pdf_file_path)
    title = pdf_document.metadata['title'].strip()
    pdf_document.close()
    return title

if __name__ == "__main__":
    # Define the root folder path
    root_folder = '/content/drive/MyDrive/NLP Pubmed'

    # Define the output file paths
    entities_file_path = '/content/drive/MyDrive/NLP Pubmed/entities.txt'
    metadata_file_path = '/content/drive/MyDrive/NLP Pubmed/entities_metadata.txt'

    # Create entity file and metadata file
    create_entity_file(root_folder, entities_file_path, metadata_file_path)
```

```python
import os
import fitz
from nltk.tokenize import word_tokenize
from PyDictionary import PyDictionary

def read_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return ""

def process_paper(pdf_path):
    doc = fitz.open(pdf_path)
    paper_text = ""

    for page_num in range(doc.page_count):
        page = doc[page_num]
        paper_text += page.get_text()

    # Tokenize the text
    words = word_tokenize(paper_text)

    return words

def create_and_translate_dictionaries(root_folder, entities_file_path):
    dictionary = PyDictionary()

    # Create a dictionary to store meanings
    entity_meanings = {}

    # Read entities from entities.txt
    with open(entities_file_path, 'r', encoding='utf-8') as entities_file:
        entities_list = [line.strip() for line in entities_file.readlines()]

    # Process each paper, extract entities, and translate meanings
    for count in range(1, 101):
        pdf_path = os.path.join(root_folder, 'Research_Papers', f'pepper_{str(count).zfill(3)}P.pdf')
        entities_in_paper = process_paper(pdf_path)

        # Filter entities based on the ones present in entities.txt
        filtered_entities = [entity for entity in entities_in_paper if entity in entities_list]

        # Translate meanings
        for entity in filtered_entities:
            if entity not in entity_meanings:
                meanings = dictionary.meaning(entity)
                if meanings:
                    entity_meanings[entity] = meanings

    # Save meanings to files
    for entity, meanings in entity_meanings.items():
        file_path = os.path.join(root_folder, f'{entity}D.txt')
        with open(file_path, 'w', encoding='utf-8') as entity_file:
            entity_file.write(f"Meanings for {entity}:\n")
            for pos, mean in meanings.items():
                entity_file.write(f"{pos.capitalize()}: {', '.join(mean)}\n")

if __name__ == "__main__":
    # Define the root folder path
    root_folder = '/content/drive/MyDrive/NLP Pubmed'
    entities_file_path = os.path.join(root_folder, 'entities.txt')

    # Create and translate dictionaries
    create_and_translate_dictionaries(root_folder, entities_file_path)
```

```python
import os
import fitz
import spacy

def read_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return ""

def process_paper(pdf_path, nlp):
    doc = fitz.open(pdf_path)
    paper_text = ""

    for page_num in range(doc.page_count):
        page = doc[page_num]
        paper_text += page.get_text()

    # Process the text using spaCy
    doc = nlp(paper_text)

    # Extract entities excluding values, dates, etc.
    filtered_entities = [ent.text for ent in doc.ents if ent.label_ not in ['DATE', 'TIME', 'PERCENT', 'MONEY', 'QUANTITY', 'ORDINAL', 'CARDINAL']]

    return filtered_entities

def create_entities_folder(root_folder, output_folder):
    nlp = spacy.load('en_core_web_sm')

    # Create the 'entities' folder
    entities_folder = os.path.join(root_folder, 'Entities')
    os.makedirs(entities_folder, exist_ok=True)

    # Process each paper and create entity files
    for count in range(1, 101):
        pdf_path = os.path.join(root_folder, 'Papers', f'olive_oil_{str(count).zfill(3)}P.pdf')
        entities_in_paper = process_paper(pdf_path, nlp)

        # Write entities to a text file
        entity_file_path = os.path.join(entities_folder, f'olive_oil_{str(count).zfill(3)}E.txt')
        with open(entity_file_path, 'w', encoding='utf-8') as entity_file:
            for entity in entities_in_paper:
                entity_file.write(f"{entity}\n")

    print(f'Entities folder created at: {entities_folder}')

if __name__ == "__main__":
    # Define the root folder path
    root_folder = '/content/drive/MyDrive/NLP Pubmed'

    # Create the 'entities' folder and store entity files
    create_entities_folder(root_folder, 'entities')
```

```python
import os
import spacy
from google.colab import drive

def read_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return ""

def create_entity_dictionary(root_folder):
    nlp = spacy.load('en_core_web_sm')

    # Create the 'entities' folder path
    entities_folder = os.path.join(root_folder, 'Entities')

    # Create an empty dictionary to store meanings
    word_meanings = {}

    # Process each entity file and extract meanings
    for count in range(1, 101):
        entity_file_path = os.path.join(entities_folder, f'pepper_{str(count).zfill(3)}E.txt')
        entities = read_text_file(entity_file_path).split('\n')

        for entity in entities:
            if entity:
                # Process the entity text using spaCy
                doc = nlp(entity)

                # Extract the lemma and its context (a few surrounding words)
                lemma = doc[0].lemma_
                context = ' '.join([token.text for token in doc])

                # Update the dictionary with the lemma and its context
                if lemma not in word_meanings:
                    word_meanings[lemma] = []

                word_meanings[lemma].append(context)

    return word_meanings

def save_dictionary_to_file(dictionary, output_file_path):
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        for lemma, contexts in dictionary.items():
            for context in contexts:
                output_file.write(f"{lemma}: {context}\n")

if __name__ == "__main__":
    # Mount Google Drive
    drive.mount('/content/drive')

    # Define the root folder path
    root_folder = '/content/drive/MyDrive/NLP Pubmed'

    # Create the entity dictionary
    entity_dictionary = create_entity_dictionary(root_folder)

    # Define the output file path for the dictionary
    dictionary_output_path = os.path.join(root_folder, 'dictionary.txt')

    # Save the dictionary to the output file
    save_dictionary_to_file(entity_dictionary, dictionary_output_path)

    print(f'Dictionary created and saved to: {dictionary_output_path}')
```

```python
import os
import spacy
from google.colab import drive
from PyDictionary import PyDictionary

def read_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return ""

def get_word_meaning(word):
    dictionary = PyDictionary()
    try:
        meaning = dictionary.meaning(word)
        if meaning:
            return meaning
        else:
            return f"No meaning found for {word}"
    except Exception as e:
        print(f"Error fetching meaning for {word}: {e}")
        return f"Error fetching meaning for {word}"

from PyDictionary import PyDictionary

def create_entity_dictionary(root_folder):
    nlp = spacy.load('en_core_web_sm')
    dictionary = PyDictionary()

    # Create the 'entities' folder path
    entities_folder = os.path.join(root_folder, 'Entities')

    # Create empty dictionaries to store English and Tamil meanings
    english_meanings = {}

    # Process each entity file and extract meanings
    for count in range(1, 101):
        entity_file_path = os.path.join(entities_folder, f'pepper_{str(count).zfill(3)}E.txt')
        entities = read_text_file(entity_file_path).split('\n')

        for entity in entities:
            if entity:
                # Process the entity text using spaCy
                doc = nlp(entity)

                try:
                    # Extract the lemma
                    if len(doc) > 0:  # Check if there's at least one token in the document
                        lemma = doc[0].lemma_

                        # Get the English meaning using PyDictionary
                        meaning = dictionary.meaning(lemma)

                        # Update the English dictionary with the lemma and its meaning
                        if lemma not in english_meanings:
                            english_meanings[lemma] = []

                        english_meanings[lemma].append(meaning)
                    else:
                        print(f"Warning: Empty document for entity in file {entity_file_path}")
                except Exception as e:
                    print(f"Error processing entity in file {entity_file_path}: {e}")

    return english_meanings



if __name__ == "__main__":
    # Mount Google Drive
    drive.mount('/content/drive')

    # Define the root folder path
    root_folder = '/content/drive/MyDrive/NLP Pubmed'

    # Create the English entity dictionary
    english_entity_dictionary = create_entity_dictionary(root_folder)

    # Define the output file path for the dictionary
    english_dictionary_output_path = os.path.join(root_folder, 'Dictionary.txt')

    # Save the English dictionary to the output file
    with open(english_dictionary_output_path, 'w', encoding='utf-8') as output_file:
        for lemma, meanings in english_entity_dictionary.items():
            for i, meaning in enumerate(meanings):
                output_file.write(f"{lemma} ({i + 1}): {meaning}\n")

    print(f'English Dictionary created and saved to: {english_dictionary_output_path}')
```
